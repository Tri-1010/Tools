```python
# Macro data with vnstock3 (robust)
import sys
import subprocess

def ensure_vnstock3():
    try:
        import vnstock3
        return
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-U", "vnstock3==3.2.1"])

ensure_vnstock3()

macro = None
using = None
try:
    from vnstock3 import Vnstock
    vn = Vnstock()
    macro = vn.macro(symbol="VN")
    using = "vnstock3.Vnstock"
except Exception as e:
    print(f"Vnstock class not available, fallback macro module: {e}")
    from vnstock3 import macro as macro_module
    macro = macro_module
    using = "vnstock3.macro"

def safe_call(fn, **kwargs):
    try:
        return fn(**kwargs)
    except Exception as exc:
        print(f"{fn.__name__} failed: {exc}")
        return None

gdp_yearly = safe_call(macro.gdp_annual, country="VN")
gdp_quarter = safe_call(macro.gdp_quarter, country="VN")
cpi = safe_call(macro.cpi, symbol="VN", source="wb")
policy_rates = safe_call(macro.policy_rate, country="VN")
fx = safe_call(macro.fx_rate, pair="USD/VND")
trade_balance = safe_call(macro.trade_balance, country="VN")

print(f"Using {using}")
if gdp_yearly is not None:
    print(gdp_yearly.head())
else:
    print("gdp_yearly is None")
```
